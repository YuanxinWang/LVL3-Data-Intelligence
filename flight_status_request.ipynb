```python
# load library
import requests
import json
from datetime import date

# set up endpoints
base_url = "https://lh-proxy.onrender.com"
password = dbutils.secrets.get(scope = "lh_secrets", key = "proxy_pass")
headers = {"password": password}
volume_path = "/Volumes/workspace/lufthansa/raw_data/"

today_date = date.today().strftime('%Y-%m-%d')
origin = "FRA"
destination = "LHR"

endpoint = f"/v1/operations/flightstatus/route/{origin}/{destination}/{today_date}"
full_url = base_url + endpoint

# send request
print(f"Requesting data for {today_date}")
response = requests.get(full_url, headers = headers)

# check response and store data
if response.status_code == 200:
    print(f"Succeeded")
    data = response.json()
    file_path = f"{volume_path}flight_status_{origin}_{destination}_{today_date}.json"
    with open(file_path, "w") as f:
        json.dump(data, f)
    print(f"Data saved in: {file_path}")

# in case of failing
else:
    print(f"Error for {today_date} flight status. Code: {response.status_code}")
    print(f"Error Message: {response.text}")
```
