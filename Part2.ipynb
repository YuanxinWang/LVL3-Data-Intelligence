```python
# load library
import requests
import json
import time

# set up endpoints
base_url = "https://lh-proxy.onrender.com"
password = dbutils.secrets.get(scope = "lh_secrets", key = "proxy_pass")
headers = {"password": password}
volume_path = "/Volumes/workspace/lufthansa/raw_data/"

reference_endpoints = {
    "airports": "/v1/references/airports",
    "airlines": "/v1/references/airlines",
    "aircraft": "/v1/references/aircraft",
    "cites": "/v1/references/cites",
    "countries": "/v1/references/countires"
}

# sending request
print("Sending API Request...")
for name, endpoint in reference_endpoints.items():
    full_url = base_url + endpoint
    print(f"Requesting data for {name}...")
    response = requests.get(full_url, headers=headers)

    # checking result and store data
    if response.status_code == 200:
        print("Succeeded")
        data = response.json()
        file_path = f"{volume_path}{name}.json"
        
        with open(file_path, "w") as file:
            json.dump(data, file)
        print(f"Data saved in: {file_path}")

    # in case of failing
    else:
        print(f"Error for {name}. Code: {response.status_code}")
        print(f"Error Message: {response.text}")
```
